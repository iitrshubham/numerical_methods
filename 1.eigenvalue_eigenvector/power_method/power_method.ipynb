### Function to calculate rank, eigen value and eigen vector of a matrix

```python
def eigen(m):
    w,v = np.linalg.eig(m)
    R = np.linalg.matrix_rank(m)
    print("rank of matrix",(R))
    print('eigen values:',w)
    print('eigen vectors:',v)
    print('-----------------------------------------------------------')
```

### Function to find largest eigen value and eigen vector using power method

```python
def power(a,x):
    print('-----------------------------------------------------------')
    print ('Power Method to find largest eigen value and eigen vector')

    n = int(len(a))

    tolerable_error = float(input('Enter tolerable error: '))


    max_iteration = int(input('Enter the number of iterations: '))
    
    # Power Method Implementation
    lambda_old = 1.0
    condition =  True
    step = 1
    while condition:
        # Multiplying a and x
        x = np.matmul(a,x)

        # Finding new Eigen value and Eigen vector
        lambda_new = max(abs(x))

        x = x/lambda_new

        # Displaying Eigen value and Eigen Vector
        print('\nIteration %d' %(step))
        print('-------------------')
        print('Maximum Eigen Value = %0.4f' %(lambda_new))
        print('Maximum Eigen Vector: ')
        for i in range(n):
            print('%0.3f\t' % (x[i]))

        # Checking maximum iteration
        step = step + 1
        if step > max_iteration:
            print('The given number of iteration is less than the maximum iteration.')
            break

        # Calculating error
        error = abs(lambda_new - lambda_old)
        print('error='+ str(error))
        lambda_old = lambda_new
        condition = error > tolerable_error
```
