<h> Gauss Seidel Method <h><br>
    <br>
    It is one of the iterative method used for solving the systems of linear equations and the basic theory which is required to implement it in python coding is presented below

<h> Notes on stopping criterion <h> <br><br>
For stopping the iteration with some convergene criterion, L infinity norm is used and it<br>
is the metric that represents the greatest length or size of a vector or matirx component. 

```python
import numpy as np
def gauss_seidel(A, b, tolerance=1e-20, max_iterations=4):
    
    x = np.zeros_like(b, dtype=np.double)
    
    
    #Iterate
    for k in range(max_iterations):
        
        x_old=x.copy() 
        
        #Loop over rows
        for i in range(A.shape[0]):
            x[i] = (b[i] - np.dot(A[i,:i], x[:i]) - np.dot(A[i,(i+1):], x_old[(i+1):])) / A[i ,i]
            print(k, x[0], x[1], x[2])
        #Stop condition 
        if np.linalg.norm(x - x_old, ord=np.inf) / np.linalg.norm(x, ord=np.inf) < tolerance:
            break
            
    return x
```

```python
A = np.array([[3,-0.1,-0.2],[0.1,7,-0.3],[0.3,-0.2,10]])
b = np.array([7.85,-19.3,71.4])
x = gauss_seidel(A, b)

print('Gauss seidel result is x = \n %s' % x)
```
